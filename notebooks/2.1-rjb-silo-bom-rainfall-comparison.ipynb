## Compare BOM monthly station data with SILO monthy station data.

* BOM monthly rainfall data is for actual observations
* SILO monthly rainfall data interpolates missing observations with observations from nearby stations
* percentile analysis with SILO interpolated data seems to be giving too high of percentiles for July throughout central wheatbelt
* Will do comparison of the two datasets



```python
import pandas as pd
```

```python
# July rainfall observations from BOM. Contains 207 SILO database stations
df_bom_july = pd.read_csv(r'C:\Users\rj71b\geo-projects\wheatbelt_rainfall_analyser\data\interim\bom_201907.csv',
                         usecols = [0,3])
```

```python
# July rainfall observations from SILO database. Contains 335 staions
df_silo_july = pd.read_csv(r'C:\Users\rj71b\geo-projects\wheatbelt_rainfall_analyser\data\interim\new_silo_weather_july.csv',
                          usecols = [2,6])
```

```python
df_silo_july.shape
```

```python
df_merge = pd.merge(df_silo_july, df_bom_july, on = "station", how = 'inner',
         suffixes=('_silo', '_bom'))
```

```python
df_merge
```

```python
# 77 stations that have different rainfall observations
df_merge[(df_merge['rain_silo'] != df_merge['rain_bom'])]
```

```python
# add new column with percent difference between observations
df_merge['percent_diff'] = df_merge.apply(lambda row: (row.iloc[2]-row.iloc[0])/row.iloc[2]*100, axis=1)
```

```python
# filter stations where diffence between rainfall observation is greater than 10 percent or less than -10 percent
#36 station where obsevation is greater than 10 percent
df_percent = df_merge[(df_merge['percent_diff'] > 10) | (df_merge['percent_diff'] < -10)]
```
